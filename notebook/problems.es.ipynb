```python
import random
from collections import Counter
import matplotlib.pyplot as plt
import math
```

```python
n = 1000                                           # Numero de veces que se lanza el dado
favorables = 0
sumas = []
p_acum = []

for t in range(1, n+1):                            # Simulacion de ambos dados
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    s = dado1 + dado2
    sumas.append(s)

    if (s > 7) or (s % 2 == 0):                     # Veces que suma mayor a 7 o un numero par
        favorables += 1

    p_acum.append(favorables / t)

print("Casos favorables:", favorables)
print("Probabilidad estimada:", favorables / n)
```

Probabilidad con 3 bolas blancas y 2 bolas rojas.

```python
iteraciones = 1000
bolas = (["White"] * 10) + (["Red"] * 20) + (["Green"] * 30)

conteo_3b_2r = 0
prob_acumulada_3b_2r = []

for t in range(1, iteraciones + 1):
    muestra = [random.choice(bolas) for _ in range(5)]

    blancas = muestra.count("White")
    rojas = muestra.count("Red")

    if blancas == 3 and rojas == 2:
        conteo_3b_2r += 1

    prob_acumulada_3b_2r.append(conteo_3b_2r / t)

print("Probabilidad estimada (3 blancas y 2 rojas):", conteo_3b_2r / iteraciones)

plt.figure()
plt.plot(range(1, iteraciones + 1), prob_acumulada_3b_2r)
plt.title("Convergencia: P(3 blancas y 2 rojas)")
plt.xlabel("N칰mero de simulaciones")
plt.ylabel("Probabilidad estimada (acumulada)")
plt.show()
```

### Combinatoria (combinaciones)
Es contar cu치ntas formas hay de elegir un grupo de cosas cuando **el orden NO importa**.  
Piensa: **라rmar un equipo**. Ana+Luis es el mismo equipo que Luis+Ana.

F칩rmula:
C(n,k) = n! / (k! (n-k)!)

### Permutaci칩n
Es contar cu치ntas formas hay de ordenar cosas cuando **el orden S칈 importa**.  
Piensa: **랃acer una fila / podio**. 1.췈 Ana, 2.췈 Luis NO es igual que 1.췈 Luis, 2.췈 Ana.

### 쮺칩mo se relaciona con probabilidad?
La probabilidad casi siempre es:
P(evento) = casos_favorables / casos_posibles

Combinatoria y permutaci칩n se usan para **contar** esos casos sin listar todo:
- Si el orden **no importa** (mano, grupo)  usa **combinatoria**.
- Si el orden **s칤 importa** (primero/segundo, secuencia)  usa **permutaci칩n**.

**Regla de oro:** si cambiar el orden no cambia el resultado  combinatoria; si cambia  permutaci칩n.

Diferencia entre combinatoria y permutaci칩n.
</h3>

La diferencia clave entre **combinatoria** y **permutaci칩n** est치 en si el **orden** cambia o no el resultado. En **combinatoria (combinaciones)** t칰 est치s contando cu치ntas formas hay de **elegir un grupo** de elementos, y el orden no importa: escoger {Ana, Luis, Sof칤a} es exactamente lo mismo que {Luis, Sof칤a, Ana}, porque al final el grupo es el mismo. En cambio, en **permutaci칩n** est치s contando cu치ntas formas hay de **ordenar** esos elementos o **asignar posiciones** (como primero, segundo, tercero), y ah칤 el orden s칤 cambia el resultado: ㄹna 1.췈, Luis 2.췈 no es lo mismo que Luis 1.췈, Ana 2.췈. Por eso, cuando el problema habla de **puestos, ranking, secuencias, contrase침as o 칩rdenes**, normalmente es permutaci칩n; cuando habla de **equipos, manos de cartas, seleccionar personas o formar grupos**, normalmente es combinatoria. La regla pr치ctica es: si al intercambiar el orden obtienes un resultado 띿iferente, es permutaci칩n; si sigue siendo 랇o mismo, es combinatoria.
