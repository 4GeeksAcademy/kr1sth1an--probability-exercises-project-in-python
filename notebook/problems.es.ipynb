    Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulaci√≥n para encontrar la probabilidad estimada
    de que la puntuaci√≥n sea mayor a 7 o un n√∫mero par.

  <div style="margin-top:12px; font-size:16px; opacity:.92; line-height:1.55;">

      <li>Ejecuta el experimento <b>1000</b> veces (lanza 2 dados 1000 veces y suma el n√∫mero de ambos dados).</li>
      <li>Lleva una cuenta de los n√∫meros y las veces que la suma fue <b>mayor a 7</b> o un <b>n√∫mero par</b>.</li>
      <li>Divide el n√∫mero del paso 2 entre el n√∫mero de iteraciones (<b>1000</b>).</li>
    </ul>
  </div>

```python
import random
from collections import Counter
import matplotlib.pyplot as plt
```

```python
n = 1000                                           # Numero de veces que se lanza el dado
favorables = 0
sumas = []
p_acum = []

for t in range(1, n+1):                            # Simulacion de ambos dados
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    s = dado1 + dado2
    sumas.append(s)

    if (s > 7) or (s % 2 == 0):                     # Veces que suma mayor a 7 o un numero par
        favorables += 1

    p_acum.append(favorables / t)

print("Casos favorables:", favorables)
print("Probabilidad estimada:", favorables / n)
```

Probabilidad con 3 bolas blancas y 2 bolas rojas.
</h3>

```python
iteraciones = 1000
bolas = (["White"] * 10) + (["Red"] * 20) + (["Green"] * 30)

conteo_3b_2r = 0
prob_acumulada_3b_2r = []

for t in range(1, iteraciones + 1):
    muestra = [random.choice(bolas) for _ in range(5)]

    blancas = muestra.count("White")
    rojas = muestra.count("Red")

    if blancas == 3 and rojas == 2:
        conteo_3b_2r += 1

    prob_acumulada_3b_2r.append(conteo_3b_2r / t)

print("Probabilidad estimada (3 blancas y 2 rojas):", conteo_3b_2r / iteraciones)

plt.figure()
plt.plot(range(1, iteraciones + 1), prob_acumulada_3b_2r)
plt.title("Convergencia: P(3 blancas y 2 rojas)")
plt.xlabel("N√∫mero de simulaciones")
plt.ylabel("Probabilidad estimada (acumulada)")
plt.show()
```

Pero conforme haces m√°s simulaciones, la l√≠nea se va estabilizando y termina cerca de un valor fijo. Eso significa que tu estimaci√≥n ya es confiable.

Probablidad de tomar todas del mismo color
</h3>

```python
iteraciones = 1000
bolas = (["White"] * 10) + (["Red"] * 20) + (["Green"] * 30)

conteo_mismo_color = 0
prob_acumulada_mismo_color = []

for t in range(1, iteraciones + 1):
    muestra = [random.choice(bolas) for _ in range(5)]

    blancas = muestra.count("White")
    rojas = muestra.count("Red")
    verdes = muestra.count("Green")

    if blancas == 5 or rojas == 5 or verdes == 5:
        conteo_mismo_color += 1

    prob_acumulada_mismo_color.append(conteo_mismo_color / t)

print("Probabilidad estimada (5 del mismo color):", conteo_mismo_color / iteraciones)

plt.figure()
plt.plot(range(1, iteraciones + 1), prob_acumulada_mismo_color)
plt.title("Convergencia: P(5 del mismo color)")
plt.xlabel("N√∫mero de simulaciones")
plt.ylabel("Probabilidad estimada (acumulada)")
plt.show()
```
