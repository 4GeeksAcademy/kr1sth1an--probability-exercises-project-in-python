```python
import random
```

```python
n = 1000
favorables = 0

for t in range(n):
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    suma = dado1 + dado2

    # Evento: suma mayor a 7 o un numero par
    if (suma > 7) or (suma % 2 == 0):
        favorables += 1

prob_estimada = favorables / n

print("Casos favorables:", favorables)
print("Probabilidad estimada:", prob_estimada)
```

```python
ball_box = {}

# Crea la caja con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"
```

```python
iteraciones = 1000
cont_3b_2r = 0
llaves = list(ball_box.keys())

for _ in range(iteraciones):
    blancas = 0
    rojas = 0

    for _ in range(5):
        color = ball_box[random.choice(llaves)]
        if color == "White":
            blancas += 1
        elif color == "Red":
            rojas += 1

    if blancas == 3 and rojas == 2:
        cont_3b_2r += 1

prob_3b_2r = cont_3b_2r / iteraciones
print("Probabilidad estimada (3 Blancas y 2 Rojas):", prob_3b_2r)
```

```python
cont_mismo_color = 0

for _ in range(iteraciones):
    blancas = 0
    rojas = 0
    verdes = 0

    for _ in range(5):
        color = ball_box[random.choice(llaves)]
        if color == "White":
            blancas += 1
        elif color == "Red":
            rojas += 1
        else:
            verdes += 1

    if blancas == 5 or rojas == 5 or verdes == 5:
        cont_mismo_color += 1

prob_mismo_color = cont_mismo_color / iteraciones
print("Probabilidad estimada (5 del mismo color):", prob_mismo_color)
```
