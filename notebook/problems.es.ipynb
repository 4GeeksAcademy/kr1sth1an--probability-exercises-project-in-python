    Se lanzan dos dados una vez y se observa la suma total. Usa una simulaci√≥n para encontrar la probabilidad estimada de que la suma sea par o mayor que 7. Una simulaci√≥n consiste en repetir el mismo experimento muchas veces para observar su comportamiento
  </div>

  <div style="margin-top:12px; font-size:16px; opacity:.92; line-height:1.55;">

      <li>Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces y suma los n√∫meros de ambos dados).</li>
      <li>Lleva el conteo de cu√°ntas veces la suma fue mayor que 7 o un n√∫mero par.</li>
      <li>Divide el n√∫mero del paso 2 entre el n√∫mero de iteraciones (1000).</li>
    </ul>
  </div>

</div>

```python
import random
from collections import Counter
import matplotlib.pyplot as plt
import math
```

```python
n = 1000                                           # Numero de veces que se lanza el dado
favorables = 0
sumas = []
p_acum = []

for t in range(1, n+1):                            # Simulacion de ambos dados
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    s = dado1 + dado2
    sumas.append(s)

    if (s > 7) or (s % 2 == 0):                     # Veces que suma mayor a 7 o un numero par
        favorables += 1

    p_acum.append(favorables / t)                   # Division entre el numero de iteraciones

print("Casos favorables:", favorables)
print("Probabilidad estimada:", favorables / n)
```

Probabilidad con 3 bolas blancas y 2 bolas rojas.

```python
iteraciones = 1000
bolas = (["White"] * 10) + (["Red"] * 20) + (["Green"] * 30)

conteo_3b_2r = 0
prob_acumulada_3b_2r = []

for t in range(1, iteraciones + 1):
    muestra = [random.choice(bolas) for _ in range(5)]

    blancas = muestra.count("White")
    rojas = muestra.count("Red")

    if blancas == 3 and rojas == 2:
        conteo_3b_2r += 1

    prob_acumulada_3b_2r.append(conteo_3b_2r / t)

print("Probabilidad estimada (3 blancas y 2 rojas):", conteo_3b_2r / iteraciones)

plt.figure()
plt.plot(range(1, iteraciones + 1), prob_acumulada_3b_2r)
plt.title("Convergencia: P(3 blancas y 2 rojas)")
plt.xlabel("N√∫mero de simulaciones")
plt.ylabel("Probabilidad estimada (acumulada)")
plt.show()
```

Probablidad de tomar todas del mismo color

```python
iteraciones = 1000
bolas = (["White"] * 10) + (["Red"] * 20) + (["Green"] * 30)

conteo_mismo_color = 0
prob_acumulada_mismo_color = []

for t in range(1, iteraciones + 1):
    muestra = [random.choice(bolas) for _ in range(5)]

    blancas = muestra.count("White")
    rojas = muestra.count("Red")
    verdes = muestra.count("Green")

    if blancas == 5 or rojas == 5 or verdes == 5:
        conteo_mismo_color += 1

    prob_acumulada_mismo_color.append(conteo_mismo_color / t)

print("Probabilidad estimada (5 del mismo color):", conteo_mismo_color / iteraciones)

plt.figure()
plt.plot(range(1, iteraciones + 1), prob_acumulada_mismo_color)
plt.title("Convergencia: P(5 del mismo color)")
plt.xlabel("N√∫mero de simulaciones")
plt.ylabel("Probabilidad estimada (acumulada)")
plt.show()
```

Qu√© es combinatoria, permutaci√≥n y como esto se relaciona con probabilidad.
</h3>


### Combinatoria (combinaciones)
Es contar cu√°ntas formas hay de elegir un grupo de cosas cuando **el orden NO importa**.  
Piensa: **‚Äúarmar un equipo‚Äù**. Ana+Luis es el mismo equipo que Luis+Ana.

<br><span style="font-size:28px; color:#ff0000;">
$C(n,k)=\frac{n!}{k!(n-k)!}$
</span>

### Permutaci√≥n
Es contar cu√°ntas formas hay de ordenar cosas cuando **el orden S√ç importa**.  
Piensa: **‚Äúhacer una fila / podio‚Äù**. 1.¬∫ Ana, 2.¬∫ Luis NO es igual que 1.¬∫ Luis, 2.¬∫ Ana.

### ¬øC√≥mo se relaciona con probabilidad?
La probabilidad casi siempre es:

<br><span style="font-size:28px; color:#ff0000;">
$P(evento) = casos_favorables / casos_posibles$
</span>

Combinatoria y permutaci√≥n se usan para **contar** esos casos sin listar todo:
- Si el orden **no importa** (mano, grupo) ‚Üí usa **combinatoria**.
- Si el orden **s√≠ importa** (primero/segundo, secuencia) ‚Üí usa **permutaci√≥n**.

Diferencia entre combinatoria y permutaci√≥n.
</h3>

La diferencia clave entre **combinatoria** y **permutaci√≥n** est√° en si el **orden** cambia o no el resultado. En **combinatoria (combinaciones)** t√∫ est√°s contando cu√°ntas formas hay de **elegir un grupo** de elementos, y el orden no importa: escoger {Ana, Luis, Sof√≠a} es exactamente lo mismo que {Luis, Sof√≠a, Ana}, porque al final el grupo es el mismo. En cambio, en **permutaci√≥n** est√°s contando cu√°ntas formas hay de **ordenar** esos elementos o **asignar posiciones** (como primero, segundo, tercero), y ah√≠ el orden s√≠ cambia el resultado: ‚ÄúAna 1.¬∫, Luis 2.¬∫‚Äù no es lo mismo que ‚ÄúLuis 1.¬∫, Ana 2.¬∫‚Äù. Por eso, cuando el problema habla de **puestos, ranking, secuencias, contrase√±as o √≥rdenes**, normalmente es permutaci√≥n; cuando habla de **equipos, manos de cartas, seleccionar personas o formar grupos**, normalmente es combinatoria. La regla pr√°ctica es: si al intercambiar el orden obtienes un resultado ‚Äúdiferente‚Äù, es permutaci√≥n; si sigue siendo ‚Äúlo mismo‚Äù, es combinatoria.
